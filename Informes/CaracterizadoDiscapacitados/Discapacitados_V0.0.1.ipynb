```python
import pandas as pd
```

```python
df = pd.read_excel('./INPUT/CONSOLIDADO.xlsx', engine='openpyxl')
```

```python
df['DETALLE_DISCAPACIDAD'].fillna(0, inplace = True)
```

```python
dis = df[(df['DISCAPACIDAD'] != 'No') | (df['DETALLE_DISCAPACIDAD'] != 0)]
```

```python
def addFranjas(franjas, diccionario):
    for franja in franjas:
        if franja in diccionario:
            diccionario[franja] += 1
        else:
            diccionario[franja] = 1
def iniciarDiccionario():
    return {
    'AUDITIVA': {
        
    },
    'VISUAL':{
        
    },
    'OTROS': {
        
    }
}
```

```python
asignaturas = {}
columnas = ['DISP_PROGRAMACION', 'DISP_INGLES', 'DISP_COACHING']
for index, row in dis.iterrows():    
    for columna in columnas:
        if not (columna in asignaturas):
            asignaturas[columna] = iniciarDiccionario()
        franjas = row[columna].split(';')
        detalle = row['DETALLE_DISCAPACIDAD']
        if isinstance(detalle, str):            
            detalles = row['DETALLE_DISCAPACIDAD'].split(';')
        else:
            detalles = []
        if 'Auditiva' in detalles:
            addFranjas(franjas, asignaturas[columna]['AUDITIVA'])
        elif 'Visual' in detalles:
            addFranjas(franjas, asignaturas[columna]['VISUAL'])
        else:
            addFranjas(franjas, asignaturas[columna]['OTROS'])
```

```python
msg = "{jornada} tiene {cantidad} solicitudes, correspondiendo a {porcentaje:.2f}%"
for asignatura, discapacidades in asignaturas.items():
    print ('#======================#')
    print (asignatura)
    for discapacidad, franjas in discapacidades.items():        
        print ('----')
        print (discapacidad)
        total = 0
        for cantidad in franjas.values():
            total += cantidad
        for franja, cantidad in franjas.items():
            print (msg.format(jornada = franja, cantidad = cantidad, porcentaje = cantidad/total*100))
        print ('', total)
    print ('\n')
```
