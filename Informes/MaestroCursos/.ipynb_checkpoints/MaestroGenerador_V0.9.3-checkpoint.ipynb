```python
import pandas as pd
from datetime import date
from numpy import int64
```

# PARÁMETROS Y CONFIGURACIÓN

```python
#Configuración
RUTA_BASE_INPUT = './INPUT/'
NOMBRE_CONSULTA_SA = 'CONSULTA_SQL.csv'
NOMBRE_CONSOLIDADO = 'CONSOLIDADO_TODOS_CICLO_3_2022-07-31.xlsx'
NOMBRE_BD_MINTIC = 'MINTIC.xlsx'
NOMBRE_BD_DOCENTES = 'INFO_DOCENTES.xlsx'
NOMBRE_BD_TUTORES = 'INFO_TUTORES.xlsx'
NOMBRE_BD_MENTORES = 'INFO_MENTORES.xlsx'
ASIGNATURAS = ['DESARROLLO_DE_SOFTWARE', 'INGLES_III', 'COACH_III'] # Poner los nombres de las asignaturas en mayúscula y en el orden del reporte del SA
NUMERO_GRUPOS = 64
# Si se deja en False, no creará todos los archivos de listas
GUARDAR_LISTAS = True
# Si se deja como False, solo guardará las listas del grupo 1 (Programación), si está activado el guardado de listas
LISTAS_TODOS_GRUPOS = True
CODIGOS_GRUPOS = ['O', 'S', 'T'] # El primer código debe ser el del grupo 1
RUTA_1DRV = 'F:/_Misión TIC 2022/OneDrive - Universidad Industrial de Santander/Documentos/Monitoría/MONITORÍA PERIODOS ACADÉMICOS/TUTORES/PERIODO 3/{tutor}/LISTAS DE CLASE/'
INCLUIR_TUTORES = True
INCLUIR_MENTORES = True


# Mapa
COLUMNAS = {
    'LLAVE_MINTIC': 'MINTIC',
    'CODIGO_UIS': 'ESTUDIANTE',
    'TIPO_DE_DOCUMENTO_ESTUDIANTE': 'TIPO DOCUMENTO',
    'NUMERO_DE_DOCUMENTO': 'DOCUMENTO',
    'APELLIDOS': 'NOMBRES',
    'NOMBRES': 'APELLIDOS',
    'EMAIL': 'EMAIL',
    'TELEFONO_SISTEMA_ACADEMICO': 'TELEFONO',
    'ASIGNATURA': 'ASIGNATURA 1',
    'GRUPO': 'GRUPO 1',
    'DOCENTE_' + ASIGNATURAS[0]: 'DOCENTE',
    'HORA_INICIAL': 'HORA INICIAL',
    'DURACION': 'DURACION',
    # INGLES
    'ASIGNATURA2': 'ASIGNATURA 2',
    'GRUPO2': 'GRUPO 2',
    'DOCENTE_'  + ASIGNATURAS[1]: 'DOCENTE.1',
    'HORA_INICIAL2': 'HORA INICIAL.1',
    'DURACION2': 'DURACION.1',
    # COACH
    'ASIGNATURA3': 'ASIGNATURA 3',
    'GRUPO3': 'GRUPO 3',
    'DOCENTE_'  + ASIGNATURAS[2]: 'DOCENTE.2',
    'HORA_INICIAL3': 'HORA INICIAL.2',
    'DIA': 'DIA',
    'DURACION3': 'DURACION.2'
}
COLUMNAS_ADICIONALES = {
    'pssBiblio': 'PSSW_BIBLIO',
    'hIniCad': 'HORA_INICIAL_',
    'hFin': 'HORA_FINAL',
    'hFinCad': 'HORA_FINAL_',
    'hIniCad2': 'HORA_INICIAL2_',
    'hFin2': 'HORA_FINAL2',
    'hFinCad2': 'HORA_FINAL2_',
    'hIniCad3': 'HORA_INICIAL3_',
    'hFin3': 'HORA_FINAL3',
    'hFinCad3': 'HORA_FINAL3_',
    'telForm': 'TELEFONO_FORMULARIO',
    'telMintic': 'TELEFONO_MINTIC',
    'nivelA': 'NIVEL_EDUCATIVO'
}
if INCLUIR_TUTORES:
    COLUMNAS_ADICIONALES['cedTutor'] = 'TUTOR'
    COLUMNAS_ADICIONALES['nTutor'] = 'NOMBRE_TUTOR'
    COLUMNAS_ADICIONALES['cTutor'] = 'CORREO_TUTOR'
    COLUMNAS_ADICIONALES['telTutor'] = 'TELEFONO_TUTOR'

if INCLUIR_MENTORES:
    COLUMNAS_ADICIONALES['cedMentor'] = 'MENTOR'
    COLUMNAS_ADICIONALES['nMentor'] = 'NOMBRE_MENTOR'
    COLUMNAS_ADICIONALES['cMentor'] = 'CORREO_MENTOR'
    #COLUMNAS_ADICIONALES['telMentor'] = 'TELEFONO_MENTOR'
# Columnas adicionales de docente
for i in range(len(ASIGNATURAS)):
    indice = str(i + 1) if i != 0 else ''
    COLUMNAS_ADICIONALES['nDAsignatura' + indice] = 'NOMBRE_DOCENTE_' + ASIGNATURAS[i]
    COLUMNAS_ADICIONALES['cDAsignatura' + indice] = 'CORREO_DOCENTE_' + ASIGNATURAS[i]
    COLUMNAS_ADICIONALES['cDIAsignatura' + indice] = 'CORREO_INSTITUCIONAL_DOCENTE_' + ASIGNATURAS[i]
    COLUMNAS_ADICIONALES['telDAsignatura' + indice] = 'TELEFONO_DOCENTE_' + ASIGNATURAS[i]

ORDEN = [
    'CODIGO_UIS',
    'NUMERO_DE_DOCUMENTO',
    'PSSW_BIBLIO',
    'NOMBRES',
    'APELLIDOS',
    'EMAIL',
    'ASIGNATURA',
    'GRUPO',
    'HORA_INICIAL',
    'HORA_INICIAL_',
    'HORA_FINAL',
    'HORA_FINAL_',
    'DURACION',
    'ASIGNATURA2',
    'GRUPO2',
    'HORA_INICIAL2',
    'HORA_INICIAL2_',
    'HORA_FINAL2',
    'HORA_FINAL2_',
    'DURACION2',
    'ASIGNATURA3',
    'GRUPO3',
    'DIA',
    'HORA_INICIAL3',
    'HORA_INICIAL3_',
    'HORA_FINAL3',
    'HORA_FINAL3_',
    'DURACION3',
    'LLAVE_MINTIC',
    'TELEFONO_MINTIC',
    'TELEFONO_FORMULARIO',
    'TELEFONO_SISTEMA_ACADEMICO',
    'TIPO_DE_DOCUMENTO_ESTUDIANTE',
    'DOCENTE_' + ASIGNATURAS[0],
    'DOCENTE_' + ASIGNATURAS[1],
    'DOCENTE_' + ASIGNATURAS[2],
    'NIVEL_EDUCATIVO'    
]    
    
for asignatura in ASIGNATURAS:
    ORDEN.append('NOMBRE_DOCENTE_' + asignatura)
    ORDEN.append('CORREO_DOCENTE_' + asignatura)
    ORDEN.append('CORREO_INSTITUCIONAL_DOCENTE_' + asignatura)
    ORDEN.append('TELEFONO_DOCENTE_' + asignatura)

if INCLUIR_TUTORES:
    ORDEN += [
        'TUTOR',
        'NOMBRE_TUTOR',
        'CORREO_TUTOR',
        'TELEFONO_TUTOR'
    ]
if INCLUIR_MENTORES:
    ORDEN += [
        'MENTOR',
        'NOMBRE_MENTOR',
        'CORREO_MENTOR',
        #'TELEFONO_MENTOR'
    ]    
    
DIAS = [
    'Lunes',
    'Martes',
    'Miércoles',
    'Jueves',
    'Viernes'
]
```

# LECTURA Y TRATAMIENTO

```python
# Lectura
consulta = pd.read_csv(RUTA_BASE_INPUT + NOMBRE_CONSULTA_SA, sep=';', encoding='latin-1', skiprows=1)
inscritos = pd.read_excel(RUTA_BASE_INPUT + NOMBRE_BD_MINTIC, engine='openpyxl')
consolidado = pd.read_excel(RUTA_BASE_INPUT + NOMBRE_CONSOLIDADO, engine='openpyxl')
docentes = pd.read_excel(RUTA_BASE_INPUT + NOMBRE_BD_DOCENTES, engine='openpyxl')
if INCLUIR_TUTORES:
    tutores = pd.read_excel(RUTA_BASE_INPUT + NOMBRE_BD_TUTORES, engine='openpyxl')
if INCLUIR_MENTORES:
    mentores = pd.read_excel(RUTA_BASE_INPUT + NOMBRE_BD_MENTORES, engine='openpyxl')
```

```python
consulta[COLUMNAS['TELEFONO_SISTEMA_ACADEMICO']] = consulta[COLUMNAS['TELEFONO_SISTEMA_ACADEMICO']].fillna(0).astype(int64, errors='ignore').astype(str)
inscritos['TELEFONO'] = inscritos['TELEFONO'].fillna(0).astype(int64, errors='ignore').astype(str)
consolidado['TELEFONO_FORMULARIO'] = consolidado['TELEFONO_FORMULARIO'].fillna(0).astype(int64, errors='ignore').astype(str)
docentes['NUMERO TELEFONICO'] = docentes['NUMERO TELEFONICO'].fillna(0).astype(int64, errors='ignore').astype(str)
if INCLUIR_TUTORES:
    tutores['Telefono'] = tutores['Telefono'].fillna(0).astype(int64, errors='ignore').astype(str)
# if INCLUIR_MENTORES:
#     mentores['Telefono'] = mentores['Telefono'].fillna(0).astype(int64, errors='ignore').astype(str)
```

## Funciones

```python
# Funciones
def buscarEnDf(df, columnaKey, key, columna):
    query = df[df[columnaKey] == key][columna]
    if len(query) > 0:        
        return query.iloc[0]
    if key == 0:
        return 'NO APLICA'
    return -1
def genBibPass(documento):
    documento = str(documento)
    return documento[len(documento) - 4:]
```

```python
if INCLUIR_TUTORES:
    tokenizados = []
    for index, row in tutores.iterrows():
        tokens = row['Consolidado grupos'].split(',')
        nTokens = []
        for token in tokens:
            token = token.replace(' ', '')
            if token != '':
                nTokens.append(token)
        tokenizados.append(nTokens)
    tutores['tokens'] = pd.Series(tokenizados)
```

# PROCESAMIENTO DE ARCHIVO MAESTRO

```python
# Procesamiento
listas = {}
RECURRENCIA = len(ASIGNATURAS)
for columna in COLUMNAS_ADICIONALES:
    listas[columna] = []
consultaE = consulta.copy()
for index, row in consultaE.iterrows():
    codigoEst = row[COLUMNAS['CODIGO_UIS']]
    listas['pssBiblio'].append(genBibPass(row[COLUMNAS['NUMERO_DE_DOCUMENTO']]))
    for i in range(1, RECURRENCIA + 1):
        indice = str(i) if i != 1 else ''
        listas['hIniCad' + indice].append(str(row[COLUMNAS['HORA_INICIAL' + indice]]) + ':00:00')
        listas['hFin' + indice].append(str(row[COLUMNAS['HORA_INICIAL' + indice]] + row[COLUMNAS['DURACION' + indice]]))
        listas['hFinCad' + indice].append(str(row[COLUMNAS['HORA_INICIAL' + indice]] + row[COLUMNAS['DURACION' + indice]]) + ':00:00')
        aDocente = ASIGNATURAS[i - 1]
        cedDocente = row[COLUMNAS['DOCENTE_' + aDocente]]
        listas['nDAsignatura' + indice].append(buscarEnDf(docentes, 'CEDULA', cedDocente, 'NOMBRE AJUSTADO'))
        listas['cDAsignatura' + indice].append(buscarEnDf(docentes, 'CEDULA', cedDocente, 'CORREO ELECTRONICO'))
        listas['cDIAsignatura' + indice].append(buscarEnDf(docentes, 'CEDULA', cedDocente, 'CUENTA_EMAIL_UIS_DEFINITIVA'))
        listas['telDAsignatura' + indice].append(buscarEnDf(docentes, 'CEDULA', cedDocente, 'NUMERO TELEFONICO'))
    consultaE.at[index, COLUMNAS['DIA']] = DIAS[row[COLUMNAS['DIA']] - 1]
    telForm = buscarEnDf(consolidado, 'CODIGO', codigoEst, 'TELEFONO_FORMULARIO')
    telMintic = buscarEnDf(inscritos, 'NUMERO', row[COLUMNAS['NUMERO_DE_DOCUMENTO']], 'TELEFONO')
    nivelA = buscarEnDf(consolidado, 'CODIGO', codigoEst, 'NIVEL_EDUCATIVO_AGRUPADO')
    listas['nivelA'].append(nivelA if nivelA != -1 else 'DESCONOCIDO')
    listas['telForm'].append(telForm if telForm != -1 else '')
    listas['telMintic'].append(telMintic if telMintic != -1 else '')
    if INCLUIR_TUTORES:
        for indext, rowt in tutores.iterrows():
            if row[COLUMNAS['GRUPO']] in rowt['tokens']:
                listas['cedTutor'].append(rowt['Cedula'])
                listas['nTutor'].append(rowt['Nombre'])
                listas['cTutor'].append(rowt['Correo'] + ';' + rowt['CorreoP'])
                listas['telTutor'].append(rowt['Telefono'])
                break
    if INCLUIR_MENTORES:
        for indexm, rowm in mentores.iterrows():
            if row[COLUMNAS['GRUPO']] == rowm['GRUPO']:
                listas['cedMentor'].append(rowm['CÉDULA MENTOR'])
                listas['nMentor'].append(rowm['NOMBRE MENTOR'])
                listas['cMentor'].append(rowm['EMAIL MENTOR'])
                #listas['telMentor'].append(rowm['Telefono'])
                break
    
for columnaNueva, lista in listas.items():
    consultaE[columnaNueva] = pd.Series(lista)
```

### Generación y guardado

```python
# Renombre
columnasAuxiliares = {v: k for k, v in COLUMNAS.items()}
columnasRenombradas = {**columnasAuxiliares, **COLUMNAS_ADICIONALES}
consultaERen = consultaE.rename(columns = columnasRenombradas)
# Ordenamiento
consultaERenOrdenada = consultaERen[ORDEN]
# Guardado
consultaERenOrdenada.to_excel('./OUTPUT/HorariosEstudiantes_C3_' + str(date.today()) + '.xlsx', index = False)
```

## PROCESAMIENTO DE ESTADÍSTICAS Y LISTAS DE CLASE

```python
nivelesEducativos = consultaERenOrdenada['NIVEL_EDUCATIVO'].dropna().unique()
estadisticasCursos = [
    ('CODIGO_CURSO', []),
    ('DOCENTE', []),
    ('NOMBRE_DOCENTE', []),
    ('CORREO_DOCENTE', []),
    ('TELEFONO_DOCENTE', []),
]
if INCLUIR_TUTORES:
    estadisticasCursos += [
        ('TUTOR', []),
        ('NOMBRE_TUTOR', []),
        ('CORREO_TUTOR', []),
        ('TELEFONO_TUTOR', []),
    ]
if INCLUIR_MENTORES:
    estadisticasCursos += [
        ('MENTOR', []),
        ('NOMBRE_MENTOR', []),
        ('CORREO_MENTOR', []),
        #('TELEFONO_MENTOR', []),
    ]
estadisticasCursos += [
    ('TOTAL', []),    
    ('INDICE_HOMOGENEIDAD', []),
]
estadisticasCursos = dict(estadisticasCursos)
for nivel in nivelesEducativos:
    estadisticasCursos[nivel] = []
limSup = 2 if not LISTAS_TODOS_GRUPOS else RECURRENCIA + 1
for grupo in range(1, limSup):    
    for i in range(1, NUMERO_GRUPOS + 1):
        iGrupo = str(grupo) if grupo != 1 else ''
        indice = str(i)
        nombreCurso = CODIGOS_GRUPOS[grupo - 1] + str(i)
        df = consultaERenOrdenada[consultaERenOrdenada['GRUPO' + iGrupo] == nombreCurso]
        if len(df) == 0:
            continue
        estadisticasCursos['CODIGO_CURSO'].append(nombreCurso)
        aDocente = ASIGNATURAS[grupo - 1]
        
        estadisticasCursos['DOCENTE'].append(df['DOCENTE_' + aDocente].iloc[0])
        estadisticasCursos['NOMBRE_DOCENTE'].append(df[COLUMNAS_ADICIONALES['nDAsignatura' + iGrupo]].iloc[0])
        estadisticasCursos['CORREO_DOCENTE'].append(df[COLUMNAS_ADICIONALES['cDAsignatura' + iGrupo]].iloc[0])
        estadisticasCursos['TELEFONO_DOCENTE'].append(df[COLUMNAS_ADICIONALES['telDAsignatura' + iGrupo]].iloc[0])
        
        if INCLUIR_TUTORES:
            estadisticasCursos['TUTOR'].append(df[COLUMNAS_ADICIONALES['cedTutor']].iloc[0] if grupo == 1 else 'NO APLICA')
            estadisticasCursos['NOMBRE_TUTOR'].append(df[COLUMNAS_ADICIONALES['nTutor']].iloc[0] if grupo == 1 else 'NO APLICA')
            estadisticasCursos['CORREO_TUTOR'].append(df[COLUMNAS_ADICIONALES['cTutor']].iloc[0] if grupo == 1 else 'NO APLICA')
            estadisticasCursos['TELEFONO_TUTOR'].append(df[COLUMNAS_ADICIONALES['telTutor']].iloc[0] if grupo == 1 else 'NO APLICA')
        
        if INCLUIR_MENTORES:
            estadisticasCursos['MENTOR'].append(df[COLUMNAS_ADICIONALES['cedMentor']].iloc[0])
            estadisticasCursos['NOMBRE_MENTOR'].append(df[COLUMNAS_ADICIONALES['nMentor']].iloc[0])
            estadisticasCursos['CORREO_MENTOR'].append(df[COLUMNAS_ADICIONALES['cMentor']].iloc[0])
            #estadisticasCursos['TELEFONO_MENTOR'].append(df[COLUMNAS_ADICIONALES['telMentor']].iloc[0])        
        
        maxN = -1
        totalCurso = len(df)
        estadisticasCursos['TOTAL'].append(totalCurso)
        for nivel in nivelesEducativos:
            cantNivel = len(df[df[COLUMNAS_ADICIONALES['nivelA']] == nivel])
            estadisticasCursos[nivel].append(cantNivel)
            if cantNivel > maxN:
                maxN = cantNivel
        estadisticasCursos['INDICE_HOMOGENEIDAD'].append(maxN / totalCurso)
        if GUARDAR_LISTAS:
            df.to_excel('./OUTPUT/CURSOS/' + nombreCurso + '.xlsx', index = False)
            if INCLUIR_TUTORES and (grupo - 1 == 0):
                tutor = df['NOMBRE_TUTOR'].iloc[0]
                rutaT = RUTA_1DRV.format(tutor = tutor)
                df.to_excel(rutaT + nombreCurso + '.xlsx', index = False)
pd.DataFrame(estadisticasCursos).to_excel('./OUTPUT/INFORME_GENERAL_CURSOS_' + str(date.today()) + '.xlsx', index = False)
```
