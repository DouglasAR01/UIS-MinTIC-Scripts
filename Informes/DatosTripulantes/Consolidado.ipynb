```python
import pandas as pd
from datetime import date
```

```python
DATOS = pd.read_excel('./HorariosEstudiantes_C4_2022-10-06.xlsx', engine = 'openpyxl')
DATOS_ANT = pd.read_excel('./datosViejos.xlsx', engine = 'openpyxl')
DATOS_NEW = pd.read_excel('./datosNuevos.xlsx', engine = 'openpyxl').drop_duplicates(subset = ['Código de estudiante'], keep = 'last')
DATOS_DEP = pd.read_excel('./datosTripulantes2.xlsx', engine = 'openpyxl')

GRUPOS_EDUCATIVOS = {
    'BACHILLER': [
        'Ninguno',
        'Primaria',
        'Bachillerato técnico',
        'Bachillerato académico',
        'Bachillerato comercial'
    ],
    'TÉCNICO': [
        'Técnica',
        'Tecnológico'
    ],
    'PROFESIONAL+': [
        'Posgrado',
        'Profesional'
    ]
}
```

```python
def ingresarDatos(df, nombreCol, encontrado):
    if not encontrado:
        return 'DESCONOCIDO'
    
    dato = df[nombreCol].iloc[0]    
    if dato == None or dato == '':
        return 'DESCONOCIDO'
    
    return dato
def checkNivelEdu(niv):
    for grupoE in GRUPOS_EDUCATIVOS:
        for nivel in GRUPOS_EDUCATIVOS[grupoE]:
            if niv == nivel:
                return grupoE
    return 'ERROR'
```

```python
rta = {
    'DOCUMENTO': [],
    'CODIGO': [],
    'NOMBRE': [],
    'APELLIDO': [],
    'TELEFONO': [],
    'EMAIL': [],
    'FECHA_NACIMIENTO': [],
    'GENERO': [],
    'NIVEL_EDUCATIVO_REAL': [],
    'OCUPACION': [],
    'TITULO_SI_TIENE': [],
    'DISCAPACIDAD': [],
    'DETALLE_DISCAPACIDAD': [],
    'PAIS_DE_RESIDENCIA': [],
    'DEPTO_DE_RESIDENCIA': [],
    'CIUDAD_DE_RESIDENCIA': [],
    'NIVEL_EDUCATIVO_AGRUPADO': [],
}
for index, row in DATOS.iterrows():
    cod = row['CODIGO_UIS']
    # Datos reciclables
    rta['DOCUMENTO'].append(row['NUMERO_DE_DOCUMENTO'])
    rta['CODIGO'].append(cod)
    rta['NOMBRE'].append(row['NOMBRES'])
    rta['APELLIDO'].append(row['APELLIDOS'])
    rta['TELEFONO'].append(row['TELEFONO_SISTEMA_ACADEMICO'])
    rta['EMAIL'].append(row['EMAIL'])
    # Datos que se toman sí o sí del consolidado viejo
    datosViejo = DATOS_ANT[DATOS_ANT['CODIGO'] == cod]
    fViejo = True
    if len(datosViejo) != 1:
        fViejo = False
        print ('Datos anteriores no encontrados para', cod)       
    
    rta['FECHA_NACIMIENTO'].append(ingresarDatos(datosViejo, 'FECHA_NACIMIENTO', fViejo))
    rta['GENERO'].append(ingresarDatos(datosViejo, 'GENERO', fViejo))
    rta['OCUPACION'].append(ingresarDatos(datosViejo, 'OCUPACION', fViejo))
    rta['TITULO_SI_TIENE'].append(ingresarDatos(datosViejo, 'TITULO_SI_TIENE', fViejo))
    
    # Datos que se toman sí o sí del dep
    datosDep = DATOS_DEP[DATOS_DEP['CODIGO'] == cod]
    fDep = True
    if len(datosDep) != 1:
        fDep = False
        print ('Datos dep no encontrados para', cod) 
    
    rta['PAIS_DE_RESIDENCIA'].append(ingresarDatos(datosDep, 'PAIS', fDep))
    rta['DEPTO_DE_RESIDENCIA'].append(ingresarDatos(datosDep, 'DEPARTAMENTO', fDep))
    rta['CIUDAD_DE_RESIDENCIA'].append(ingresarDatos(datosDep, 'CIUDAD', fDep))
    
    # Datos que se actualizan
    datosNew = DATOS_NEW[DATOS_NEW['Código de estudiante'] == cod]
    fNew = True
    if len(datosNew) != 1:
        fNew = False
        print ('Datos new no encontrados para', cod)
    if not fNew and not fViejo:
        rta['NIVEL_EDUCATIVO_REAL'].append('DESCONOCIDO')
        rta['DISCAPACIDAD'].append('DESCONOCIDO')
        rta['DETALLE_DISCAPACIDAD'].append('DESCONOCIDO')
        rta['NIVEL_EDUCATIVO_AGRUPADO'].append('DESCONOCIDO')
    elif not fNew and fViejo:
        rta['NIVEL_EDUCATIVO_REAL'].append(ingresarDatos(datosViejo, 'NIVEL_EDUCATIVO_REAL', fViejo))
        rta['DISCAPACIDAD'].append(ingresarDatos(datosViejo, 'DISCAPACIDAD', fViejo))
        rta['DETALLE_DISCAPACIDAD'].append(ingresarDatos(datosViejo, 'DETALLE_DISCAPACIDAD', fViejo))
        rta['NIVEL_EDUCATIVO_AGRUPADO'].append(ingresarDatos(datosViejo, 'NIVEL_EDUCATIVO_AGRUPADO', fViejo))
    else: # Existe el nuevo y no el viejo o existen ambos
        nivel = ingresarDatos(datosNew, 'Nivel de escolaridad', fNew)
        rta['NIVEL_EDUCATIVO_REAL'].append(nivel)
        rta['DISCAPACIDAD'].append(ingresarDatos(datosNew, '¿Presenta alguna discapacidad?', fNew))
        rta['DETALLE_DISCAPACIDAD'].append(ingresarDatos(datosNew, "Si en la anterior pregunta su respuesta fue 'Sí', por favor especifique el tipo de discapacidad.", fNew))
        rta['NIVEL_EDUCATIVO_AGRUPADO'].append(checkNivelEdu(nivel))
```

```python
datosRta = pd.DataFrame(rta)
datosRta['GENERO'] = datosRta['GENERO'].str.upper()
datosRta['NIVEL_EDUCATIVO_REAL'] = datosRta['NIVEL_EDUCATIVO_REAL'].str.upper()
datosRta.to_excel('./DatosRelevantesCaracterizaciónC4_' + str(date.today()) + '.xlsx', index = False)
```
