```python
import pandas as pd
import glob
```

```python
archivos = glob.glob('./INPUT/SPRINTS/*.csv')
dfs = []
for archivo in archivos:
    dfs.append(pd.read_csv(archivo, encoding='utf-8'))
```

```python
datos = pd.concat(dfs)
```

```python
sprints = datos[['Número de ID',
                 'Tarea:Sprint 1. Identificando los objetivos del proyecto (Real)',
                 'Tarea:Sprint 2. Desarrollando del Backend (Real)',
                 'Tarea:Sprint 3. Desarrollando el Frontend (Real)',
                 'Tarea:Sprint 4. Desplegando la aplicación (Real)',
                 'Tarea:Sprint 1. Definiendo el proyecto (Real)',
                 'Tarea:Sprint 2. Primeros pasos en el desarrollo de la APP (Real)',
                 'Tarea:Sprint 3. Desarrollando las funciones (Real)',
                 'Tarea:Sprint 4. Finalizando la app (Real)'
                ]]
```

```python
sprints
```

```python
rta = {
    'CÓDIGO': [],
    'CICLO': [],
}
for i in range (1, 5):
    rta['SPRINT ' + str(i)] = []
for index, row in sprints.iterrows():
    rta['CÓDIGO'].append(row['Número de ID'].split('-')[1])
    if not pd.isna(row['Tarea:Sprint 1. Identificando los objetivos del proyecto (Real)']):
        rta['CICLO'].append('4A')
        rta['SPRINT 1'].append(row['Tarea:Sprint 1. Identificando los objetivos del proyecto (Real)'])
        rta['SPRINT 2'].append(row['Tarea:Sprint 2. Desarrollando del Backend (Real)'])
        rta['SPRINT 3'].append(row['Tarea:Sprint 3. Desarrollando el Frontend (Real)'])
        rta['SPRINT 4'].append(row['Tarea:Sprint 4. Desplegando la aplicación (Real)'])
    else:
        rta['CICLO'].append('4B')
        rta['SPRINT 1'].append(row['Tarea:Sprint 1. Definiendo el proyecto (Real)'])
        rta['SPRINT 2'].append(row['Tarea:Sprint 2. Primeros pasos en el desarrollo de la APP (Real)'])
        rta['SPRINT 3'].append(row['Tarea:Sprint 3. Desarrollando las funciones (Real)'])
        rta['SPRINT 4'].append(row['Tarea:Sprint 4. Finalizando la app (Real)'])
rta = pd.DataFrame(rta)
rta.replace('-', 0, inplace = True)
for i in range(1, 5):
    rta['SPRINT ' + str(i)] = rta['SPRINT ' + str(i)].astype(float)
```

```python
calc = []
res = []
peso = (0.3125, 0.3125, 0.1875, 0.1875)
for index, row in rta.iterrows():
    acum = 0
    for i in range(4):
        acum += row['SPRINT ' + str(i+1)] * peso[i]
    calc.append(acum)
    if acum < 3.7:
        res.append('SÍ')
    else:
        res.append('NO')
rta['NOTA PARCIAL'] = calc
rta['RIESGO'] = res
```

```python
rta.to_excel('./Heuristica.xlsx', index = False)
```
