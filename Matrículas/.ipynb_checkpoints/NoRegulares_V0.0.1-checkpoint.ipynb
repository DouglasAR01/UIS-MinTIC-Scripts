```python
COD_GRUPOS = {}
for asignatura, listaH in HORARIOS.items():
    if asignatura == 'DESARROLLO DE SOFTWARE':
        cod = 'O'
    elif asignatura == 'INGLES III':
        cod = 'S'
    else:
        cod = 'T'
    i = 1
    for hora, cantidad in listaH:
        for i in range(i, i + cantidad):
            COD_GRUPOS[cod + str(i)] = hora
            i += 1
```

```python
def permutar(listaProgramacion, listaIngles, listaCoach):
    res = []
    for horaInicio, horaFin in listaProgramacion:
        for horaIngles in listaIngles:
            if horaIngles < horaInicio or horaIngles > horaFin:
                for horaCoach in listaCoach:
                    if horaCoach != horaIngles and (horaCoach < horaInicio or horaCoach > horaFin):
                        res.append(((horaInicio, horaFin), horaIngles, horaCoach))
    return res
```

```python
def menorCurso(asignatura, horaCurso, tipo, cod):
    if tipo == 1:
        conjunto = TRADUCTOR_PROGRAMACION
    else:
        conjunto = TRADUCTOR_INGLES_COACH
    horaABuscar = conjunto[horaCurso]
    gruposDisponibles = []
    for codigo, horario in COD_GRUPOS.items():
        if codigo[0] == cod and horario == horaABuscar:
            gruposDisponibles.append(codigo)
    minGrupo = 100000000
    minGrupoCod = -1
    for grupo in gruposDisponibles:
        cant = conteoGrupos[asignatura][grupo]
        if cant < minGrupo and cant < 75:
            minGrupo = cant
            minGrupoCod = grupo
    if minGrupoCod == -1:
        return 'ERROR', 400000000000000000
    return minGrupoCod, minGrupo
    
def menorCostoCurso(emparejamientos):
    costoMinimo = 100000
    codigosMinimos = -1
    emparejamientoMinimo = -1
    for emparejamiento in emparejamientos:
        i = 0        
        costoActual = 0
        codigos = []
        for j in range(len(emparejamiento)):
            tipo = 2
            if j == 0:
                asig = 'DESARROLLO DE SOFTWARE'
                tipo = 1
                cod = 'O'
            if j == 1:
                asig = 'INGLES III'
                cod = 'S'
            if j == 2:
                asig = 'COACHING III'
                cod = 'T'
            horaCurso = emparejamiento[j]
            codigoCurso, costo = menorCurso(asig, horaCurso, tipo, cod)
            costoActual += costo
            codigos.append(codigoCurso)
            i += 1
        if costoActual < costoMinimo:
            costoMinimo = costoActual
            codigosMinimos = codigos
            emparejamientoMinimo = emparejamiento
    return codigosMinimos, emparejamientoMinimo
            
```

```python
conteoGrupos = {}
for asignatura in ASIGNATURAS:    
    dfAsignatura = DATOS_GRUPOS[DATOS_GRUPOS['NOMBRE ASIGNATURA'] == asignatura]
    conteoGrupos[asignatura] = dfAsignatura['GRUPO'].value_counts().sort_values(ascending = False).to_dict()
rta = {
    'CODIGO': [],
    'NOMBRE': [],
    'APELLIDO': [],
}
posibilidades = permutar(list(FRANJAS_PROGRAMACION.values()), FRANJAS_SOLO_INGLES,list(FRANJAS_INGLES_COACH.values()))
for asignatura in ASIGNATURAS:
    rta[asignatura] = []
for indexDf, row in DATOS_EST.iterrows():
    homologado = False
    if row['CODIGO_UIS'] in HOMOLOGADOS:
        homologado = True
    rta['CODIGO'].append(row['CODIGO_UIS'])
    rta['NOMBRE'].append(row['NOMBRES'])
    rta['APELLIDO'].append(row['APELLIDOS'])
    resultado, _ = menorCostoCurso(posibilidades)
    rta['DESARROLLO DE SOFTWARE'].append(resultado[0])
    rta['INGLES III'].append(resultado[1] if not homologado else 'HOMOLOGADO')
    rta['COACHING III'].append(resultado[2])
    conteoGrupos['DESARROLLO DE SOFTWARE'][resultado[0]] += 1
    conteoGrupos['INGLES III'][resultado[1]] += 1 if not homologado else 0
    conteoGrupos['COACHING III'][resultado[2]] += 1
    
#         if asignatura == 'INGLES III' and row['CODIGO_UIS'] in HOMOLOGADOS:
#             rta[asignatura].append('HOMOLOGADO')
#             continue        
#         rta[asignatura].append(balancear(asignatura))
resultado = pd.DataFrame(rta)
resultado.to_excel('./OUTPUT/NOREGULARES/MATRICULAS_NO_REGULARES_' + str(date.today()) + '.xlsx', index = False)
```

```python
rta2 = {
    'ASIGNATURA': [],
    'GRUPO': [],
    'CANTIDAD NO REGULARES': [],
    'CANTIDAD ACTIVOS': [],
    'TOTAL ESTUDIANTES': []
}
for asignatura in ASIGNATURAS:
    valores = resultado[asignatura].value_counts().sort_values(ascending = False)
    for grupo, numero in valores.iteritems():
        total = conteoGrupos[asignatura][grupo] if grupo != 'HOMOLOGADO' else 'NA'
        rta2['ASIGNATURA'].append(asignatura)
        rta2['GRUPO'].append(grupo)
        rta2['CANTIDAD NO REGULARES'].append(numero)
        rta2['CANTIDAD ACTIVOS'].append(total - numero if grupo != 'HOMOLOGADO' else 'NA')
        rta2['TOTAL ESTUDIANTES'].append(total)
pd.DataFrame(rta2).to_excel('./OUTPUT/NOREGULARES/RESUMEN_NO_REGULAES_' + str(date.today()) + '.xlsx', index = False)    
```
